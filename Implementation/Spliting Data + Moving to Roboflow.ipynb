```python
import os
import shutil
import random

# 1. Clean and Setup the "Train" structure
base_path = '/kaggle/working/roboflow_upload'
train_path = os.path.join(base_path, 'train')

if os.path.exists(base_path):
    shutil.rmtree(base_path)
os.makedirs(train_path, exist_ok=True)

'''
target_classes = ['house', 'ice cream', 'laptop', 'lion', 'mermaid', 'monkey', 'octopus', 'panda',
 'pants', 'pencil', 'penguin', 'pig', 'pineapple', 'pizza', 'rabbit', 'sea turtle',
 'shark','sheep', 'shoe', 'smiley face', 'spider', 'squirrel', 'star', 'strawberry',
 'sun','swan', 't-shirt', 'teddy-bear', 'television', 'traffic light', 'train', 'tree',
 'umbrella', 'vase', 'watermelon']



 Final classes for me = ['house', 'ice cream', 'laptop', 'lion', 'mermaid', 'monkey', 'octopus', 'panda',
 'pants', 'pencil', 'penguin', 'pineapple', 'pizza', 'rabbit', 'sea turtle',
 'sheep', 'shoe', 'smiley face', 'spider', 'squirrel', 'star', 'strawberry',
 'sun', 't-shirt', 'television', 'traffic light', 'train', 'tree',
 'umbrella', 'vase', 'watermelon']
'''
# 2. Configuration
src_root = '/kaggle/input/datasets/ashishjangra27/doodle-dataset/doodle' # Verify this path in your sidebar
target_classes = ["panda" , "squirrel" , "monkey" , "sea turtle"]
limit = 1200
# pineapple = 0.89     0.90  
# pinguin = 0.90    tohfa
# rabbit = 0.90       tohfa
# squirrel = 0.88    tohfa   

# panda = 0.89         999
# sea turtle = 0.86        940
# pig = 0.84       
# teddy-bear =  0.86


# 3. Random Selection & Copy
for cls in target_classes:
    src_cls_path = os.path.join(src_root, cls)
    dst_cls_path = os.path.join(train_path, cls)
    
    if not os.path.exists(src_cls_path):
        print(f"⚠️ Skipping {cls}: Folder not found.")
        continue
        
    os.makedirs(dst_cls_path, exist_ok=True)
    all_imgs = os.listdir(src_cls_path)
    
    # Pick 1000 random images
    selected = random.sample(all_imgs, min(len(all_imgs), limit))
    
    for img in selected:
        shutil.copy(os.path.join(src_cls_path, img), os.path.join(dst_cls_path, img))
        
    print(f"✅ Prepared {len(selected)} images for class: {cls}")
```
